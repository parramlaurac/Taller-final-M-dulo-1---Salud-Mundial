### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzavida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Promedio

esperanza_vida= [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
años = 0

for vida in esperanza_vida:
    años += vida

promedio = años / len(esperanza_vida)

print("La esperanza de vida promedio es: {:.1f}".format(promedio), "años")

# Máximo
max = esperanza_vida[0]  # empezamos suponiendo que el primero es el mayor
for valor in esperanza_vida:
    if valor > max:
        max = valor

print("El valor máximo es:", max)

# Mínimo
min = esperanza_vida[0]  # empezamos suponiendo que el primero es el menor
for valor in esperanza_vida:
    if valor < min:
        min = valor

print("El valor mínimo es:", min)
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
print("============================================")
print("      CLASIFICAR ESPERANZA DE VIDA")

vida = []

def clasificar_esperanza_vida(anios=None):
    while True:
        try:
            if anios is None:
                anios = float(input("Ingrese los años de esperanza de vida: "))
            
            if anios < 0:
                print("Debe ser un número mayor o igual a 0")
                break

            if anios >= 75:
                print("La esperanza de vida es alta")
            elif anios >= 60:
                print("La esperanza de vida es media")
            else:
                print("La esperanza de vida es baja")

            return anios  

        except ValueError:
            print("Edad inválida, ingrese solo números")


while True:
    print("\n===========================================")
    print("1. Registrar")
    print("2. Salir")

    op = input("Opción: ")

    if op == "1":
        anios = clasificar_esperanza_vida()
        vida.append({"años": anios})

    elif op == "2":
        break

    else:
        print("Opción inválida, digite 1 o 2")


if len(vida) > 0:
    print("\nAños registrados:")
    
    for p in vida:
        if p['años'] >= 75:
            tipo = "Esperanza de vida ALTA"
        elif p['años'] >= 60:
            tipo = "Esperanza de vida MEDIA"
        else:
            tipo = "Esperanza de vida BAJA"

        print(f"Edad registrada: {p['años']:.1f} años → {tipo}")

print("\nEl proceso ha finalizado, ¡Buen día!")
````

### 2.6 — Nueva columna calculada

🔧 **Ejercicio 2.6**
1. Crea una nueva columna llamada `categoria_vida` aplicando la función `clasificar_esperanza_vida` que creaste en la Parte 1 a la columna `esperanza_vida`
2. ¿Cuántos países caen en cada categoría? (usa `value_counts()`)
3. ¿Qué porcentaje del total representa cada categoría?

````python
# Tu código aquí

# Creando nueva columna "categoria_vida" aplicando la función clasificar_esperanza_vida
def clasificar_esperanza_vida(anios):
    if anios >= 75:
        return "Alta"
    elif anios >= 60:
        return "Media"
    else:
        return "Baja"  
df_nuevo["categoria_vida"] = df_nuevo["esperanza_vida"].apply(clasificar_esperanza_vida)
print("           CATEGORÍA DE ESPERANZA DE VIDA")
print("=====================================================")
print(df_nuevo[["pais", "esperanza_vida", "categoria_vida"]].head(8).round(1))
````

````python
# Cuantos paises caen en cada categoria 
paises_por_categoria = df_nuevo["categoria_vida"].value_counts()
print("NÚMERO DE PAÍSES POR CATEGORÍA DE ESPERANZA DE VIDA")
print("=====================================================")
print(paises_por_categoria) 
````

````python
# Que porcentaje de países tienen esperanza de vida alta, media y baja
total_paises = df_nuevo.shape[0]
print("PORCENTAJE DE PAÍSES POR CATEGORÍA DE ESPERANZA DE VIDA")
print("========================================================")
for categoria, cantidad in paises_por_categoria.items():
    porcentaje = (cantidad / total_paises) * 100
    print(f"{categoria}: {porcentaje:.1f}% ({cantidad} países)")
    
````
